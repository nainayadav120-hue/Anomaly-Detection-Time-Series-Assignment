```python
#1. What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
""" Anomaly Detection
Anomaly detection is the process of identifying data points or observations that deviate significantly from the expected behavior or norm of a dataset. These anomalies can indicate unusual patterns, errors, or rare events that may require further investigation.

Types of Anomalies
There are three primary types of anomalies:
1. Point AnomaliesPoint anomalies are individual data points that are significantly different from the rest of the data. These anomalies are usually easy to detect and can be identified using statistical methods.
Example:
- A temperature reading of 100°C in a dataset of normal temperatures ranging from 20°C to 30°C.
- A transaction of $10,000 in a bank account with usual transactions ranging from $10 to $100.

2. Contextual AnomaliesContextual anomalies are data points that are anomalous in a specific context or situation. These anomalies depend on the context in which they occur.
Example:
- A person's credit card transaction in a foreign country may be considered anomalous if they have never traveled abroad before.
- A sudden increase in network traffic during a holiday weekend may be normal, but the same increase on a regular weekday could be anomalous.

3. Collective AnomaliesCollective anomalies are a set of data points that are anomalous when considered together, but not individually. These anomalies often require analyzing the relationships between data points.
Example:
- A series of small transactions from the same account to different recipients may be considered anomalous if they occur within a short period.
- A group of sensors in a industrial control system reporting abnormal readings in a coordinated manner may indicate a potential security breach.

Anomaly detection is a critical task in various applications, including:
- Fraud detection
- Network security
- Predictive maintenance
- Quality control
- Medical diagnosis

By identifying anomalies, organizations can detect potential issues, prevent losses, and improve overall performance."""
```

```python
#2. Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
""" Comparison of Anomaly Detection Algorithms
This comparison highlights the approaches and suitable use cases for Isolation Forest, DBSCAN, and Local Outlier Factor (LOF) algorithms.

Isolation Forest- Approach: Isolation Forest is an ensemble method that uses multiple decision trees to isolate anomalies. It works by building trees that split the data in a way that anomalies are isolated faster than normal data points.
- Suitable Use Cases:
    - High-dimensional data
    - Large datasets
    - Data with complex structures
    - Real-time anomaly detection
- Advantages:
    - Scalable and efficient
    - Handles high-dimensional data well
    - Robust to outliers in the training data
- Disadvantages:
    - May not perform well with noisy data
    - Requires careful tuning of hyperparameters

DBSCAN- Approach: DBSCAN is a density-based clustering algorithm that groups data points into clusters based on their density. It identifies anomalies as points that do not belong to any cluster.
- Suitable Use Cases:
    - Data with varying densities
    - Data with clusters of varying sizes
    - Identifying clusters and anomalies simultaneously
- Advantages:
    - Handles varying densities well
    - Identifies clusters and anomalies simultaneously
    - Robust to noise
- Disadvantages:
    - Sensitive to parameter tuning (ε and min_samples)
    - May not perform well with high-dimensional data

Local Outlier Factor (LOF)- Approach: LOF is a density-based algorithm that calculates the local density of each point and compares it to the densities of its neighbors. Points with a significantly lower density than their neighbors are considered anomalies.
- Suitable Use Cases:
    - Data with varying densities
    - Identifying anomalies in local regions
    - Data with complex structures
- Advantages:
    - Handles varying densities well
    - Identifies local anomalies
    - No need to specify the number of clusters
- Disadvantages:
    - Sensitive to parameter tuning (k)
    - May not perform well with high-dimensional data"""
```

```python
#3. What are the key components of a Time Series? Explain each with one example.
""" Key Components of a Time Series
A time series is a sequence of data points measured at regular time intervals. The key components of a time series are:
1. Trend- Definition: The overall direction or pattern in the data over a long period of time.
- Example: A company's sales have been increasing steadily over the past 5 years, indicating a positive trend.
2. Seasonality- Definition: Regular fluctuations in the data that occur at fixed intervals, such as daily, weekly, monthly, or yearly.
- Example: A retail store experiences higher sales during the holiday season (December to January) every year, indicating a seasonal pattern.
3. Cyclical Component- Definition: Long-term fluctuations in the data that are not seasonal, but rather related to economic or business cycles.
- Example: A construction company's sales may follow the economic cycle, with more sales during periods of economic growth and fewer sales during recessions.
4. Irregular Component (Noise)- Definition: Random, unpredictable events or fluctuations in the data that are not part of the trend, seasonality, or cyclical components.
- Example: A sudden power outage at a factory may cause a one-time drop in production, which is an irregular event that is not part of the overall trend or seasonality.
5. Level- Definition: The average value of the time series over a specific period of time.
- Example: A company's average daily website traffic is 10,000 visitors, which represents the level of the time series.

Understanding these components is essential for:
- Forecasting: Identifying the trend, seasonality, and cyclical components can help predict future values.
- Anomaly detection: Identifying irregular components can help detect unusual patterns or events.
- Modeling: Understanding the components can help build more accurate models of the time series.

By decomposing a time series into its components, analysts can gain a better understanding of the underlying patterns and make more informed decisions."""
```

```python
#4. Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

"""Stationarity in Time Series
A time series is said to be stationary if its statistical properties, such as the mean, variance, and autocorrelation, remain constant over time. In other words, a stationary time series has no trend, seasonality, or other patterns that change over time.

Types of Stationarity
There are two types of stationarity:
- Weak Stationarity: The first two moments (mean and covariance) of the time series are constant over time.
- Strong Stationarity: All moments of the time series are constant over time.

Testing for Stationarity
There are several tests to determine if a time series is stationary:
- Augmented Dickey-Fuller (ADF) Test: A popular test for unit roots, which can indicate non-stationarity.
- KPSS Test: A test for trend stationarity, which can detect if a time series has a trend.

Transforming a Non-Stationary Series
To transform a non-stationary series into a stationary one, you can use the following methods:
1. Differencing- First-order differencing: Subtract each value from the previous value to remove the trend.
- Higher-order differencing: Apply differencing multiple times to remove higher-order trends.
2. Normalization- Log transformation: Take the logarithm of the time series to stabilize the variance.
- Standardization: Subtract the mean and divide by the standard deviation to normalize the time series.
3. Detrending- Linear detrending: Remove a linear trend from the time series.
- Non-linear detrending: Remove a non-linear trend from the time series.
4. Seasonal Decomposition- Seasonal differencing: Subtract the value from the same season in the previous year to remove seasonality.
- Seasonal decomposition: Use techniques like STL decomposition or seasonal decomposition to separate the time series into trend, seasonality, and residuals."""

#Example Code (Python)

import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller

# Generate a non-stationary time series
np.random.seed(0)
n = 100
time = np.arange(n)
series = np.sin(time) + 0.1 * time + np.random.randn(n)

# Perform ADF test
result = adfuller(series)
print("ADF Statistic:", result[0])
print("p-value:", result[1])

# Difference the time series
diff_series = np.diff(series)

# Perform ADF test on the differenced series
result = adfuller(diff_series)
print("ADF Statistic (after differencing):", result[0])
print("p-value (after differencing):", result[1])

""""By testing and transforming a non-stationary series into a stationary one, you can:
- Improve model performance: Many time series models assume stationarity, so transforming the data can improve model accuracy.
- Enhance interpretability: Stationary time series are easier to understand and analyze.
- Simplify forecasting: Stationary time series are often easier to forecast."""
```

```python
#5. Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
"""Time Series Models
This section compares and contrasts the structure and application of AR, MA, ARIMA, SARIMA, and SARIMAX models.
1. AR (Autoregressive) Model- Structure: AR models use past values of the time series to forecast future values.
- Equation: y_t = β0 + β1*y_(t-1) + … + βp*y_(t-p) + ε_t
- Application: Suitable for time series with strong autocorrelation, where past values are good predictors of future values.
2. MA (Moving Average) Model- Structure: MA models use the errors (residuals) from past predictions to improve future forecasts.
- Equation: y_t = ε_t + θ1*ε_(t-1) + … + θq*ε_(t-q)
- Application: Suitable for time series with strong moving average components, where errors from past predictions can help improve future forecasts.
3. ARIMA (Autoregressive Integrated Moving Average) Model- Structure: ARIMA models combine AR and MA components with differencing to handle non-stationarity.
- Equation: y_t = β0 + β1*y_(t-1) + … + βp*y_(t-p) + ε_t + θ1*ε_(t-1) + … + θq*ε_(t-q)
- Application: Suitable for time series with strong autocorrelation, moving average components, and non-stationarity.
4. SARIMA (Seasonal ARIMA) Model- Structure: SARIMA models extend ARIMA by incorporating seasonal components.
- Equation: Similar to ARIMA, but with additional seasonal terms.
- Application: Suitable for time series with strong seasonality, where the same patterns repeat at regular intervals.
5. SARIMAX (Seasonal ARIMA with Exogenous Variables) Model- Structure: SARIMAX models extend SARIMA by incorporating exogenous variables (external factors that affect the time series).
- Equation: Similar to SARIMA, but with additional terms for exogenous variables.
- Application: Suitable for time series with strong seasonality, non-stationarity, and external factors that affect the series."""
```

```python
#6. Load a time series dataset (e.g., AirPassengers), plot the original series, and decompose it into trend, seasonality, and residual components.
#(Include your Python code and output in the code box below.)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Load the AirPassengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, index_col='Month', parse_dates=['Month'])

# Plot the original series
plt.figure(figsize=(10,6))
plt.plot(df['Passengers'])
plt.title('Original Series')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()

# Decompose the series into trend, seasonality, and residual components
decomposition = seasonal_decompose(df['Passengers'], model='multiplicative')

# Plot the decomposition
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(10,8))
plt.subplot(411)
plt.plot(df['Passengers'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
#7. Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot.
#(Include your Python code and output in the code box below.)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Generate a synthetic dataset to simulate NYC Taxi Fare data
np.random.seed(42)
n_samples = 1000  # Number of regular data points
n_anomalies = 20  # Number of anomalous data points

# Generate regular data points for NYC area
pickup_longitude_normal = np.random.normal(loc=-73.98, scale=0.05, size=n_samples)
pickup_latitude_normal = np.random.normal(loc=40.75, scale=0.03, size=n_samples)
fare_amount_normal = np.random.normal(loc=12.0, scale=3.0, size=n_samples)

# Generate anomalous data points (e.g., very far locations or extremely high/low fares)
pickup_longitude_anomaly = np.random.uniform(low=-80, high=-60, size=n_anomalies) # Far longitude
pickup_latitude_anomaly = np.random.uniform(low=30, high=50, size=n_anomalies)   # Far latitude
fare_amount_anomaly = np.random.uniform(low=50.0, high=200.0, size=n_anomalies)  # High fares

# Combine normal and anomalous data
pickup_longitude = np.concatenate((pickup_longitude_normal, pickup_longitude_anomaly))
pickup_latitude = np.concatenate((pickup_latitude_normal, pickup_latitude_anomaly))
fare_amount = np.concatenate((fare_amount_normal, fare_amount_anomaly))

df = pd.DataFrame({
    'pickup_longitude': pickup_longitude,
    'pickup_latitude': pickup_latitude,
    'fare_amount': fare_amount
})

# Filter out any negative fare amounts that might have been generated by random normal
df = df[df['fare_amount'] > 0].reset_index(drop=True)

# Select relevant columns (already done during synthetic data creation)
# df = df[['pickup_longitude', 'pickup_latitude', 'fare_amount']]

# Drop rows with missing values (not applicable for synthetic data but kept for consistency if external data was used)
df.dropna(inplace=True)

# Apply Isolation Forest
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['pickup_longitude', 'pickup_latitude', 'fare_amount']])

isolation_forest = IsolationForest(contamination=0.05, random_state=42) # Increased contamination for synthetic data
isolation_forest.fit(df_scaled)
anomaly = isolation_forest.predict(df_scaled)

# Identify anomalies
df['anomaly'] = anomaly
anomalies = df[df['anomaly'] == -1]

# Visualize the anomalies
plt.figure(figsize=(12, 10))
plt.scatter(df['pickup_longitude'], df['pickup_latitude'], c=['blue' if x == 1 else 'red' for x in df['anomaly']], alpha=0.5, s=5)
plt.title('Anomalies in NYC Taxi Fare Data (Red points are anomalies)')
plt.xlabel('Pickup Longitude')
plt.ylabel('Pickup Latitude')
plt.grid(True)
plt.show()
```

```python
#8. Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results.
#(Include your Python code and output in the code box below.)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Load the dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, index_col='Month', parse_dates=['Month'])

# Plot the original series
plt.figure(figsize=(10,6))
plt.plot(df['Passengers'])
plt.title('Original Series')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.show()

# Train a SARIMA model
model = SARIMAX(df['Passengers'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()

# Forecast the next 12 months
forecast = results.forecast(steps=12)

# Plot the forecast
plt.figure(figsize=(10,6))
plt.plot(df['Passengers'], label='Original')
plt.plot(pd.date_range(start=df.index[-1] + pd.DateOffset(months=1), periods=12), forecast, label='Forecast', marker='o', linestyle='--', color='red')
plt.title('Forecast')
plt.xlabel('Month')
plt.ylabel('Passengers')
plt.legend()
plt.show()
```

```python
#10.  You are working as a data scientist for a power grid monitoring company.
#Your goal is to forecast energy demand and also detect abnormal spikes or drops in real-time consumption data collected every 15 minutes.
#The dataset includes features like timestamp, region, weather conditions, and energy usage.
#Explain your real-time data science workflow:
#● How would you detect anomalies in this streaming data (Isolation Forest / LOF / DBSCAN)?
#● Which time series model would you use for short-term forecasting (ARIMA / SARIMA / SARIMAX)?
#● How would you validate and monitor the performance over time?
#● How would this solution help business decisions or operations?
#(Include your Python code and output in the code box below.)
"""Real-Time Data Science Workflow for Energy Demand Forecasting and Anomaly Detection
As a data scientist for a power grid monitoring company, our goal is to forecast energy demand and detect abnormal spikes or drops in real-time consumption data. Here's a step-by-step approach to achieve this:

Step 1: Data Ingestion and Preprocessing
- Collect real-time energy consumption data every 15 minutes from various regions.
- Preprocess the data by handling missing values, outliers, and data normalization.
- Convert the data into a suitable format for analysis.

Step 2: Anomaly Detection
- Use Isolation Forest algorithm to detect anomalies in the energy consumption data.
- Train the model on historical data and use it to detect anomalies in real-time data.
- Identify the anomalies and alert the operations team.

Step 3: Short-Term Forecasting
- Use SARIMA (Seasonal ARIMA) model for short-term forecasting of energy demand.
- Train the model on historical data and use it to forecast energy demand for the next 24 hours.
- Use the forecasted values to optimize energy distribution and reduce waste.

Step 4: Validation and Monitoring
- Validate the performance of the anomaly detection and forecasting models using metrics such as Mean Squared Error (MSE) and Mean Absolute Error (MAE).
- Monitor the performance of the models over time and retrain them as necessary.

Step 5: Business Impact
- Use the insights from the anomaly detection and forecasting models to inform business decisions and optimize operations.
- Identify areas of inefficiency and opportunities for improvement.
- Develop strategies to reduce energy waste and improve the overall efficiency of the power grid."""

#Python Code

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error

# Generate some sample data
np.random.seed(0)
n_samples = 1000
timestamp = pd.date_range('2022-01-01', periods=n_samples, freq='15min')
energy_usage = np.random.rand(n_samples) * 100 + np.sin(np.arange(n_samples)) * 10
region = np.random.choice(['A', 'B', 'C'], size=n_samples)
weather_conditions = np.random.choice(['Sunny', 'Cloudy', 'Rainy'], size=n_samples)

# Create a DataFrame
df = pd.DataFrame({'timestamp': timestamp, 'region': region, 'weather_conditions': weather_conditions, 'energy_usage': energy_usage})

# Anomaly detection using Isolation Forest
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[['energy_usage']])
isolation_forest = IsolationForest(contamination=0.01)
anomaly = isolation_forest.fit_predict(df_scaled)

# Identify anomalies
df['anomaly'] = anomaly

# Visualize anomalies
plt.figure(figsize=(10, 6))
plt.plot(df['energy_usage'], label='Energy Usage')
plt.scatter(df.index[df['anomaly'] == -1], df['energy_usage'][df['anomaly'] == -1], c='red', label='Anomalies')
plt.legend()
plt.show()

# Short-term forecasting using SARIMA
model = SARIMAX(df['energy_usage'], order=(1,1,1), seasonal_order=(1,1,1,96))
results = model.fit()

# Forecast the next 96 time steps (24 hours)
forecast = results.forecast(steps=96)

# Visualize forecast
plt.figure(figsize=(10, 6))
plt.plot(df['energy_usage'], label='Actual')
plt.plot(pd.date_range(start=df['timestamp'].iloc[-1] + pd.DateOffset(minutes=15), periods=96, freq='15min'), forecast, label='Forecast', marker='o', linestyle='--', color='red')
plt.legend()
plt.show()

# Validate and monitor performance
mse = mean_squared_error(df['energy_usage'][-96:], results.fittedvalues[-96:])
print(f'MSE: {mse:.2f}')

"""Business Impact
The solution can help business decisions or operations in several ways:
1. Anomaly detection can identify unusual patterns in energy usage, allowing for early intervention and preventing potential issues.
2. Short-term forecasting can help optimize energy distribution and reduce waste.
3. The solution can provide insights into energy usage patterns, allowing for more informed decisions about energy efficiency and resource allocation."""
```
